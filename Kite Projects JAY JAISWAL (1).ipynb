# 2nd Table Work

```python
weighted_avg_price = df.groupby(['instrument', 'type']).sum()
```

```python
weighted_avg_price['weighted_avg_price'] = weighted_avg_price['turnover'] / weighted_avg_price['first']
```

```python
charges_summary = df.groupby(['instrument', 'type'])['total_charges'].sum()
```

```python
charges_summary = df.groupby(['instrument', 'type'])[["first",'turnover',"brokerage","stt/ctt","etc","sebi","gst","stamp_charges",'total_charges']].sum()
```

```python
result = weighted_avg_price[['weighted_avg_price']].merge(charges_summary, on=['instrument', 'type'])
```

# 3nd Table Work

```python
df['instrument']
```

```python
pnl_summary = df.groupby(['instrument', 'type'])['turnover'].sum().unstack()
```

```python
pnl_summary['Gross_PnL'] = pnl_summary['SELL'] - pnl_summary['BUY']
```

```python
charges_summary = df.groupby(['instrument'])['total_charges'].sum()
```

```python
pnl_summary['Total_Charges'] = charges_summary
pnl_summary['Net_PnL'] = pnl_summary['Gross_PnL'] - pnl_summary['Total_Charges']
```

```python
pnl_summary['%_Charges_on_Gross_PnL'] = (pnl_summary['Total_Charges'] / pnl_summary['Gross_PnL']) * 100

print(pnl_summary[['Gross_PnL', 'Total_Charges', 'Net_PnL', '%_Charges_on_Gross_PnL']])
```
